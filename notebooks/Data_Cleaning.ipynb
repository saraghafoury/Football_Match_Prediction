# Data Cleaning

The objective of this step is to improve the quality of the dataset before training machine learning models.

The following preprocessing tasks will be performed:

- Remove unnecessary columns.
- Handle inconsistent columns.
- Remove duplicate records.
- Handle missing values.

The cleaning steps are ordered to maximize data quality. Categorical values are standardized before removing duplicates so that records differing only in formatting are identified as duplicates. Missing values are then handled, followed by data type conversion and categorical encoding to prepare the dataset for machine learning.

These operations ensure that the dataset is clean, consistent, and suitable for machine learning algorithms.

## Remove Unnecessary Columns

The `match_id` column is only an identifier and does not contain predictive information.
Therefore, it is removed from the dataset.

```python
df.drop(columns=["match_id"], inplace=True)
```

## Consistency Check of Categorical Variables

The categorical variables `home_club`, `away_club`, and `competition` were inspected to identify inconsistent naming, duplicated categories, unnecessary whitespace, and typographical errors.

```python
categorical_cols = ["home_club", "away_club", "competition"]

# Count unique values BEFORE cleaning
before = {}

for col in categorical_cols:
    before[col] = df[col].nunique()

# Cleaning

# Remove unnecessary whitespace
for col in categorical_cols:
    df[col] = (
        df[col]
        .str.strip()
        .replace(r"\s+", " ", regex=True)
    )

# Correct obvious spelling mistakes
competition_mapping = {
    "Permier League Play-offs": "Premier League Play-offs"
}

df["competition"] = df["competition"].replace(competition_mapping)

# Count unique values AFTER cleaning

after = {}

for col in categorical_cols:
    after[col] = df[col].nunique()

# Compare

print("=" * 60)
print("CATEGORICAL CONSISTENCY CHECK")
print("=" * 60)

for col in categorical_cols:
    print(f"{col}")
    print(f"Before cleaning : {before[col]}")
    print(f"After cleaning  : {after[col]}")
    print(f"Corrected       : {before[col] - after[col]}")
    print("-" * 60)
```

## Remove Duplicate Records

Duplicate records may negatively affect model performance.
This step removes duplicated rows while keeping the first occurrence.

```python
initial_rows = len(df)

df.drop_duplicates(inplace=True)

print(f"Removed {initial_rows - len(df)} duplicate trip records.")
```

## Handling Missing Values

Missing values were imputed according to the meaning of each feature instead of applying a single strategy to all columns.

The following rules were used:

- **Manager IDs** were filled with `-1`, indicating that the manager information was unavailable.
- **Cup game** and **played at home** columns were filled with `0`, assuming the match was not a cup game or was not played at home when historical information was missing.
- **Goals scored** and **goals conceded** were filled with `0`, representing no recorded goals in the missing historical match.
- **Team ratings**, **opponent ratings**, and **days since previous match** were filled with the median value of each column to preserve the overall distribution.
- **Categorical features** were filled with `"Unknown"` to retain all observations while explicitly marking missing categories.

This feature-specific strategy preserves the semantic meaning of the data and avoids introducing unnecessary bias.

```python
print("=" * 60)
print("MISSING VALUES BEFORE IMPUTATION")
print(df.isnull().sum().sum())

# -----------------------------
# Column groups
# -----------------------------

# Manager IDs
manager_cols = [c for c in df.columns if "manager_id" in c]

# Binary columns
binary_cols = [c for c in df.columns
               if ("cup_game" in c) or ("played_at_home" in c)]

# Goal columns
goal_cols = [c for c in df.columns
             if ("goals_scored" in c) or ("goals_conceded" in c)]

# Rating columns
rating_cols = [c for c in df.columns
               if ("team_rating" in c) or ("opponent_rating" in c)]

# Rest-day columns
rest_cols = [c for c in df.columns
             if "days_since_match" in c]

# Text columns
categorical_cols = df.select_dtypes(include="object").columns

# -----------------------------
# Imputation
# -----------------------------

# Manager IDs
df[manager_cols] = df[manager_cols].fillna(-1)

# Binary features
df[binary_cols] = df[binary_cols].fillna(0)

# Convert binary columns to integer
for col in binary_cols:
    if df[col].dtype == object:
        df[col] = (
            df[col]
            .replace({True: 1, False: 0, "True": 1, "False": 0})
            .fillna(0)
            .astype(int)
        )
    else:
        df[col] = df[col].astype(int)

# Goals
df[goal_cols] = df[goal_cols].fillna(0)

# Ratings
for col in rating_cols:
    df[col] = df[col].fillna(df[col].median())

# Days since previous match
for col in rest_cols:
    df[col] = df[col].fillna(df[col].median())

# Text columns
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

print("\n" + "=" * 60)
print("MISSING VALUES AFTER IMPUTATION")
print(df.isnull().sum().sum())
```

```python
comp_cols = [col for col in df.columns if "recent_competition_code" in col]

competition_cols = [
    col for col in df.columns
    if "recent_competition_code" in col
]

df[competition_cols] = df[competition_cols].fillna(-1)


print("\n" + "=" * 60)
print("MISSING VALUES AFTER IMPUTATION")
print(df.isnull().sum().sum())
```
